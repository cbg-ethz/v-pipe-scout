# Run Lollipop on from within

```python
# IMPORTs
import pandas as pd

import lollipop as lollipop
```

```python
# LOAD DATA

# mutation frequencies
mutation_frequency_df = pd.read_csv("data/mutation_counts_coverage.csv")

# variant signatures
variant_signature_df = pd.read_csv("data/mutation_variant_matrix.csv")

## Show head of the data
print("mutation_frequency_df")
print(mutation_frequency_df.head())
print("")
print("variant_signature_df")
print(variant_signature_df.head())
```

```python
# Make Parameters for Kernel 

# X (pd.DataFrame): dataframe of variant definition (design matrix)
# Assuming mutations are in rows and columns are variants, see line 44 in kerneldeconv.py
# make mutations the index
X = variant_signature_df.set_index("Mutation")

# y (pd.Series): series of observed mutation frequencies
y = mutation_frequency_df["frequency"]

# dates (pd.Series): series of observation dates / convert to datetime
dates = mutation_frequency_df["sampling_date"].apply(pd.to_datetime)

# weights (pd.Series): series of weights for the observations
# None

# kernel (kernel object): object with methods to compute kernel weighting
GaussianKernel = lollipop.GaussianKernel(bandwidth=30)

# reg (regressor object): object with methods to compute the regression
reg = lollipop.RobustReg(f_scale=0.01)

# confint (confint object): object with method to compute confidence bands
confint = lollipop.NullConfint()
```

```python
#  print X, y and dates
print("X")
print(X.head())
print("")
print("X.values()")
print(X.values)
print(f"shape: {X.shape}")
print("")
print("y")
print(y.head())
print("")
print("dates")
print(dates.head())
print(f"shape dates: {dates.shape}")
```

```python
# Make Kernel 
t_kdec = lollipop.KernelDeconv(
    X=X,
    y=y,
    dates=dates,
    reg=reg,  # type: ignore
    confint=confint, # type: ignore
)
```

```python
# Run Kernel Deconvolution
t_kdec.deconv_all(
    min_tol=1e-3,
)
```
